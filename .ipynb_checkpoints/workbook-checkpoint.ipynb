## Somatic Mutations from TCGA

Read maf files. Originally downloaded from tcga website. No special access was necessary.

```python
data_dir = 'maf_files_thyroid/'
maf_files = os.listdir(data_dir)
maf = {}
for f in maf_files:
    input_file = open(data_dir+f, 'r')
    maf[f] = pd.read_csv(input_file, sep='\t', header=0, index_col=0, low_memory=True)
```

Organize pandas data frame.

```python
m = pd.concat(maf)
m = m.dropna(axis=1, how='all')
cols = ['Ncbi_Build', 'Chrom', 'Start_Position', 
         'End_Position', 'Strand', 'Reference_Allele', 
         'Tumor_Seq_Allele1', 'Tumor_Seq_Allele2',
         'Tumor_Sample_Barcode', 'Entrez_Gene_Id',
         'Variant_Classification','Variant_Type']
m = m[cols]
m = m.reset_index()
m = m[m.Entrez_Gene_Id != 0]
```

#### Adding useful columns

Boolean for both alleles being mutated

```python
def check_bi_allelic(x):
    if x[0] == x[1]:
        return True
    else:
        return False
m["Both_Alleles"] = m[['Tumor_Seq_Allele1', 'Tumor_Seq_Allele2']].apply(check_bi_allelic, axis=1)
```

Constructing 10 digit patieng barcode

```python
m['Patient_Barcode'] = m.Tumor_Sample_Barcode.str[:12]
```

Creating a numerical identifier for patients; will be helpful for visualizations.

```python
unique_patient_identifiers = list(m.Patient_Barcode.unique())
all_patient_rows = list(m.Patient_Barcode)
all_unique_identifiers = []
for p_id in all_patient_rows:
    all_unique_identifiers.append(unique_patient_identifiers.index(p_id))
m['Patient_Number'] = pd.Series(all_unique_identifiers)
```

Creating a numerical identifier for genes; will be helpful for visualizations also.

```python
unique_gene_identifier = list(m.Entrez_Gene_Id.unique())
all_entrez = list(m.Entrez_Gene_Id)
Seq_Gene_Id = []
for gene in all_entrez:
    Seq_Gene_Id.append(unique_gene_identifier.index(gene))
m['Seq_Gene_Id'] = pd.Series(Seq_Gene_Id)
```

Reducing data frame to relevant rows.

```python
mr = m[['Hugo_Symbol', 'Entrez_Gene_Id', 'Seq_Gene_Id', 'Patient_Number', 'Patient_Barcode', 'Variant_Classification', 'Both_Alleles', ]]
```

### Getting Gene Interactions from HumanNet

HumanNet: Creating a dictionary graph for all gene connections in HumanNet.

```python
edges = {}
keys = []
with open('HumanNet.v1.benchmark.txt', 'r') as human_net:
    for line in human_net:
        c = [int(x) for x in line.strip().split("\t")]
        if c[0] in keys:
            edges[c[0]].append(c[1])
        else:
            edges[c[0]] = [c[1]]
            keys.append(c[0])
        if c[1] in keys:
            edges[c[1]].append(c[0])
        else:
            edges[c[1]] = [c[0]]
            keys.append(c[1])
```

Messing around to better understand the properties of HumanNet

```python
connect = []
for key in keys:
    connect.append(len(edges[key]))
connections = {"Gene_id": pd.Series(keys),
               "Number_of_Connections": pd.Series(connect)}
connectionsdf = pd.DataFrame(connections)
connectionsdf['Number_of_Connections'].hist(bins=100)
```

Creating bfs method to search network.

```python
def find_distance(graph, start_node, end_node):
    nodes = graph.keys()
    discovered = [0 for x in range(len(graph.keys()))]
    distance = [0 for x in range(len(graph.keys()))]
    queue = [start_node]
    discovered[nodes.index(start_node)] = 1
    while queue is not empty:
        curr = queue.pop(0)
        print curr
        # check if item is end node, if yes then break
        if curr == end_node:
            print "SUCCESS", distance[nodes.index(curr)]
            return distance[nodes.index(curr)]
        # otherwise 
        else:
            for neighbor in graph[curr]:
                if discovered[nodes.index(neighbor)] == 0:
                    distance[nodes.index(neighbor)] = distance[nodes.index(curr)] + 1
                    queue.extend(graph[neighbor])
                    discovered[nodes.index(neighbor)] = 1
```

## Mutation Classification

Creating a data frame with all mutations occuring only once. Add number of mutations per gene.

```python
mutations = pd.DataFrame(mr.Entrez_Gene_Id.value_counts())
mutations = mutations.reset_index()
mutations.columns = ['Entrez_Gene_Id', 'counts']
mutations.head()
```

Add percentage of bi-allelic mutations per gene. Add percentage of gene mutations that are missense

```python
bi_all, missense, total = [], [], []
for gene in mutations.Entrez_Gene_Id:
    bi_all.append(mr[(mr['Entrez_Gene_Id'] == gene) & mr['Both_Alleles'] == True].count()[0])
    missense.append(mr[(mr['Entrez_Gene_Id'] == gene) & (mr['Variant_Classification'] == 'Missense_Mutation')].count()[0])
    total.append(mr[mr['Entrez_Gene_Id'] == gene].count()[0])
```

```python
print bi_all[:10]
print total[:10]
print missense[:10]
mutations['Perc_Bi_Allelic'] = pd.Series(bi_all)/pd.Series(total)
mutations['Perc_Missense'] = pd.Series(missense)/pd.Series(total)
mutations.head()
```
